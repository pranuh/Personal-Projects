```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
# Define the URL you want to scrape
url = 'https://www.hdfcbank.com/personal/pay/cards/credit-cards'

# Set up headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Make the GET request with headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML with BeautifulSoup
    bs = BeautifulSoup(response.text, 'html.parser')

    # Find all the 'span' elements with the class 'cardparent' (or adjust based on actual HTML)
    titles = bs.findAll('div',{'class':'card-offer-contr'})
    
    save_data = []
    # Print the extracted text
    for title in titles:
        save_data.append(title.get_text())
else:
    print(f"Error {response.status_code}: Unable to fetch the webpage.")
```

```python
df = pd.DataFrame(save_data)
split_df = df[0].str.split('\n',expand=True)
filtered_df = split_df[[4,25]]
filtered_df
```
